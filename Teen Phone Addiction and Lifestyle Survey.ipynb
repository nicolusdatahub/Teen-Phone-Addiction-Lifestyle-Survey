```python
import pandas as pd
import os

# Dataset path (update if needed)
dataset_path = r'C:\Users\Admin\.cache\kagglehub\datasets\khushikyad001\teen-phone-addiction-and-lifestyle-survey\versions\1'

# Find the CSV file
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.endswith('.csv'):
            file_path = os.path.join(root, file)
            df = pd.read_csv(file_path)
            print(f"Loaded: {file_path}")
            break

# Preview
df.head()
```

```python
# Drop columns with more than 30% missing values
threshold = len(df) * 0.3
df.dropna(axis=1, thresh=threshold, inplace=True)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of daily phone usage
sns.histplot(df['Daily_Usage_Hours'], kde=True, color='skyblue')
plt.title("Distribution of Daily Phone Usage")
plt.xlabel("Daily Usage (hours)")
plt.show()
```

```python
# Compare screen time on weekdays vs weekends
sns.boxplot(data=df[['Daily_Usage_Hours', 'Weekend_Usage_Hours']])
plt.title("Daily vs Weekend Phone Usage")
plt.ylabel("Usage (Hours)")
plt.xticks([0, 1], ['Weekday', 'Weekend'])
plt.show()
```

```python
# Top usage purposes
sns.countplot(y='Phone_Usage_Purpose', data=df, order=df['Phone_Usage_Purpose'].value_counts().index)
plt.title("Primary Phone Usage Purpose")
plt.xlabel("Count")
plt.ylabel("Purpose")
plt.show()
```

```python
sns.scatterplot(x='Daily_Usage_Hours', y='Sleep_Hours', data=df)
plt.title("Daily Phone Usage vs Sleep Hours")
plt.xlabel("Phone Usage (hours)")
plt.ylabel("Sleep Hours")
plt.show()
```

```python
sns.scatterplot(x='Daily_Usage_Hours', y='Academic_Performance', data=df)
plt.title("Phone Usage vs Academic Performance")
plt.xlabel("Daily Usage (hours)")
plt.ylabel("Academic Score")
plt.show()
```

```python
print(df.columns.tolist())
```

```python
# Define bins and labels (adjust thresholds as needed)
bins = [0, 3, 6, 10]
labels = ['Low', 'Moderate', 'High']

df['Addiction_Category'] = pd.cut(df['Addiction_Level'], bins=bins, labels=labels, include_lowest=True)

sns.boxplot(x='Addiction_Category', y='Anxiety_Level', data=df)
plt.title("Addiction Category vs Anxiety Level")
plt.xlabel("Phone Addiction Category")
plt.ylabel("Anxiety Level")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.boxplot(x='Addiction_Level', y='Anxiety_Level', data=df, ax=axes[0])
axes[0].set_title("Addiction Level vs Anxiety")
axes[0].set_xlabel("Phone Addiction Level")
axes[0].set_ylabel("Anxiety Level")

sns.boxplot(x='Addiction_Level', y='Depression_Level', data=df, ax=axes[1])
axes[1].set_title("Addiction Level vs Depression")
axes[1].set_xlabel("Phone Addiction Level")
axes[1].set_ylabel("Depression Level")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.boxplot(x='Gender', y='Daily_Usage_Hours', data=df, ax=axes[0])
axes[0].set_title("Daily Phone Usage by Gender")
axes[0].set_xlabel("Gender")
axes[0].set_ylabel("Daily Usage (hours)")

sns.boxplot(x='Gender', y='Sleep_Hours', data=df, ax=axes[1])
axes[1].set_title("Sleep Hours by Gender")
axes[1].set_xlabel("Gender")
axes[1].set_ylabel("Sleep Hours")

plt.tight_layout()
plt.show()
```
